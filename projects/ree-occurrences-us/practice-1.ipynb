```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('C:/Users/User/Documents/geopandas-coures-files/USGS_REE_US_CSV/Loc_Pt.csv')
df.head(2)
```

```python
mineral_loc = df[['Ftr_Name', 'Ftr_Type', 'Commodity', 'Lat_WGS84', 'Long_WGS84']]
mineral_loc.head()
```

```python
mineral_loc = mineral_loc.rename(columns={'Ftr_Name':'Location', 'Ftr_Type':'Type', 'Lat_WGS84': 'Latitude', 'Long_WGS84':'Longitude'})
mineral_loc.head()
```

```python
gdf = gpd.GeoDataFrame(mineral_loc,
                       geometry=gpd.points_from_xy(mineral_loc['Longitude'], mineral_loc['Latitude']))
```
